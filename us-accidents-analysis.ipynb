```python
df.isna() #checks for missing values#
```

```python
df.isna().sum().sort_values(ascending=False) #total number of instances of missing data per column#
```

```python
missing_percentages = df.isna().sum().sort_values(ascending=False) / len(df)
```

```python
missing_percentages
```

```python
missing_percentages[ missing_percentages != 0 ] #using the boolean as an index to find only those columns which have missing data#
```

```python
missing_percentages[ missing_percentages != 0 ].plot(kind='barh')
```

**Note:** If large chunks of a particular column are missing, then the column may be irrelevant to the data analysis project. Remove the columns that you do not wish to use.
